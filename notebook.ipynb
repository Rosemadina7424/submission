## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Analisis distribusi jumlah penyewaan sepeda
plt.figure(figsize=(10, 6))
sns.histplot(df['cnt'], bins=30, kde=True)
plt.title('Distribusi Penyewaan Sepeda Harian')
plt.show()

# Menghapus kolom yang tidak relevan untuk analisis korelasi
df_numeric = df.select_dtypes(include=[np.number])  # Mengambil kolom numerik

# Analisis korelasi
plt.figure(figsize=(10, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Matriks Korelasi Variabel')
plt.show()
```

**Insight:**
- Terdapat hubungan positif yang kuat antara jumlah penyewaan sepeda dengan variabel cuaca dan musim.
- Kelembaban dan kecepatan angin tampak memiliki pengaruh negatif terhadap jumlah penyewaan.

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Visualisasi pengaruh musim terhadap jumlah penyewaan sepeda
plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='cnt', data=df)
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Musim')
plt.xticks([0, 1, 2, 3], ['Spring', 'Summer', 'Fall', 'Winter'])
plt.show()

# Visualisasi pengaruh cuaca terhadap jumlah penyewaan sepeda
plt.figure(figsize=(10, 6))
sns.boxplot(x='weathersit', y='cnt', data=df)
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Cuaca')
plt.xticks([0, 1, 2, 3], ['Clear', 'Mist', 'Light Rain/Snow', 'Heavy Rain/Snow'])
plt.show()
```

### Pertanyaan 2:

```python
# Visualisasi pengaruh hari kerja terhadap penyewaan sepeda
plt.figure(figsize=(10, 6))
sns.barplot(x='workingday', y='cnt', data=df)
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Hari Kerja vs Akhir Pekan')
plt.xticks([0, 1], ['Akhir Pekan/Hari Libur', 'Hari Kerja'])
plt.show()
```

**Insight:**
- Penyewaan sepeda tertinggi terjadi pada musim panas dan terendah di musim dingin.
- Cuaca cerah mendorong penyewaan lebih tinggi, sementara cuaca buruk seperti hujan atau salju menurunkan penyewaan.
- Jumlah penyewaan sepeda lebih tinggi pada akhir pekan dibandingkan dengan hari kerja.

## Analisis Lanjutan (Opsional)

```python
# Analisis regresi linier untuk prediksi penyewaan sepeda berdasarkan variabel cuaca
from sklearn.linear_model import LinearRegression

# Model sederhana untuk memprediksi jumlah penyewaan sepeda
X = df[['temp', 'hum', 'windspeed']]
y = df['cnt']

model = LinearRegression()
model.fit(X, y)

# Prediksi dan visualisasi hasil
predictions = model.predict(X)
plt.scatter(y, predictions)
plt.xlabel('Actual Rentals')
plt.ylabel('Predicted Rentals')
plt.title('Prediksi vs Realitas Penyewaan Sepeda')
plt.show()
```

## Conclusion

- Conclution pertanyaan 1 : Cuaca cerah dan musim panas cenderung meningkatkan jumlah penyewaan sepeda. Sebaliknya, cuaca buruk dan musim dingin menurunkan penyewaan.
- Conclution pertanyaan 2 : Penyewaan sepeda lebih tinggi pada akhir pekan dan hari libur dibandingkan hari kerja, menunjukkan bahwa orang lebih sering menggunakan sepeda untuk rekreasi saat waktu senggang.
